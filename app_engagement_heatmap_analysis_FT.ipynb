```python
import pandas as pd

file = pd.read_csv('../../Data/app_eng_18.csv')
full_df = file[['impressions', 'feature_value', 'page_type_value', 'banner_image', 'banner_landing_page', 'clicks']]
full_df
```

```python
def getsuitablefontcolor(rgb):
    R, G, B = rgb

    brightness = 0.299 * R + 0.587 * G + 0.114 * B

    threshold = 128  # Adjust this threshold as needed

    if brightness > threshold:
        return (0, 0, 0)  # It's a lighter color, return black
    else:
        return (255, 255, 255)
```

```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
import numpy as np


def hextorgb(hex):
    return tuple(int(hex[i:i+2], 16) for i in (1, 3, 5))

def annotate(path, info):
    image = Image.open(path)
    draw = ImageDraw.Draw(image)

    for feature, pos, impressions, color in info:
        box_position = (550, pos) 
        box_size = (160, 50)    
        box_color = hextorgb(color)

        draw.rectangle([box_position, (box_position[0] + box_size[0], box_position[1] + box_size[1])], fill=box_color)

        text = str(impressions)
        font = ImageFont.load_default()
        font = font.font_variant(size=40)
        text_color = getsuitablefontcolor(box_color)

        text_position = (box_position[0], box_position[1])  # Adjust the padding as needed

        draw.text(text_position, text, fill=text_color, font=font)
    image.show()
    return

def assignheatmap(features, data):
    min_data = min(data)
    max_data = max(data)

    cmap = plt.get_cmap('inferno')

    normalized_data = [(imp - min_data) / (max_data - min_data) for imp in data]

    heatmap = {}

    for i in range(len(normalized_data)):
        norm_imp = normalized_data[i]
        feature = features[i]
        color = cmap(norm_imp)
        hex_color = "#{:02x}{:02x}{:02x}".format(int(color[0] * 255), int(color[1] * 255), int(color[2] * 255))
        heatmap[feature] = hex_color

    return heatmap
```

```python
def getinfo(page, heatmap, data):
    # feature, pos
    info = []
    for feat, pos in page:
        info.append((feat, pos, data[feat], heatmap[feat]))
    return info
```

<h2>For FTs</h2>

```python
# Process for FT Landing page
df = full_df[full_df['page_type_value'] == '25469']
df
```

```python
features = [5002, 5001, 5004, 6030, 5020, 6003, 20003, 6021, 20122, 5007, 20101, 20102, 20111, 20103, 20112, 20104, 5019, 20121, 20105, 20113, 20106, 20107, 5006, 6050, 6051, 6053, 20114, 5014, 20115, 20116, 20117, 20108, 5012, 20109, 20110, 20118, 20119, 20120, 5017, 6005]

pages = [[(5002, 200),(5001, 400), (5004, 1200)],
[(6030, 160), (6005, 400), (6003, 800), (20003, 1200)],
[(6021, 300), (20122, 800), (5007, 1030), (20101, 1200)],
[(20102, 350), (20103, 700)]]
```

```python
clicks_per_feature = {}
clicks = []
for feature in features:
    click = df[df['feature_value'] == feature]['clicks'].sum()
    clicks_per_feature[feature] = click
    clicks.append(click)
```

```python
# annotate for clicks
heatmap_clicks = assignheatmap(features, clicks)

annotate("App Screenshots/FT/1.jpeg", getinfo(pages[0], heatmap_clicks, clicks_per_feature))
annotate("App Screenshots/FT/2(1).jpeg", getinfo(pages[1], heatmap_clicks, clicks_per_feature))
annotate("App Screenshots/FT/3.jpeg", getinfo(pages[2], heatmap_clicks, clicks_per_feature))
annotate("App Screenshots/FT/4.jpeg", getinfo(pages[3], heatmap_clicks, clicks_per_feature))
```

```python
impressions_per_feature = {}
impressions = []
for feature in features:
    imp = df[df['feature_value'] == feature]['impressions'].max()
    impressions_per_feature[feature] = imp
    impressions.append(imp)
```

```python
# annotate for impressions
heatmap_impressions = assignheatmap(features, impressions)

annotate("App Screenshots/FT/1.jpeg", getinfo(pages[0], heatmap_impressions, impressions_per_feature))
annotate("App Screenshots/FT/2(1).jpeg", getinfo(pages[1], heatmap_impressions, impressions_per_feature))
annotate("App Screenshots/FT/3.jpeg", getinfo(pages[2], heatmap_impressions, impressions_per_feature))
annotate("App Screenshots/FT/4.jpeg", getinfo(pages[3], heatmap_impressions, impressions_per_feature))
```
